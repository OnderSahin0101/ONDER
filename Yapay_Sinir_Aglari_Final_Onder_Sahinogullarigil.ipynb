```python
# =========================
# ISIC 2018 Binary Classification End-to-End Pipeline (TF/Keras)
# Single-Block Code (Scratch CNN + MobileNetV2 TL + EfficientNetB0 TL + Test Metrics + Grad-CAM)
# =========================

import os, random, numpy as np, pandas as pd
import tensorflow as tf
from pathlib import Path

from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score, roc_auc_score,
                             confusion_matrix, classification_report, roc_curve)

import matplotlib.pyplot as plt
import itertools

# -------------------------
# 0) SETTINGS
# -------------------------
SEED = 42
random.seed(SEED); np.random.seed(SEED); tf.random.set_seed(SEED)

AUTOTUNE  = tf.data.AUTOTUNE
IMG_SIZE  = (224, 224)
BATCH_SIZE = 32

# >>>>>> CHANGE THIS <<<<<<
# Root folder should look like: ROOT_DIR/class_name/*.jpg (or png)
ROOT_DIR = "/content/data/ROOT_DIR/Test"   # <-- CHANGE ME
# -------------------------

def build_df_from_folder(root_dir):
    root = Path(root_dir)
    if not root.exists():
        raise FileNotFoundError(f"ROOT_DIR not found: {root_dir}")

    rows = []
    for cls_dir in root.iterdir():
        if cls_dir.is_dir():
            cls = cls_dir.name
            for p in cls_dir.rglob("*"):
                if p.suffix.lower() in [".jpg", ".jpeg", ".png"]:
                    rows.append({"path": str(p), "label": cls})
    df = pd.DataFrame(rows)
    if df.empty:
        raise ValueError("No images found. Check ROOT_DIR and folder structure (root/class/*.jpg).")
    return df

# -------------------------
# 0.1) LOAD DF + TOP-2 CLASSES (adjusted for single class)
# -------------------------
df = build_df_from_folder(ROOT_DIR)
print("Total images found:", len(df))
print("\nAll class counts:\n", df["label"].value_counts())

class_counts = df["label"].value_counts()
unique_classes_in_data = class_counts.index.tolist()

is_single_class = False # Flag to track if it's a single class scenario

if len(unique_classes_in_data) < 2:
    is_single_class = True
    print(f"\nWARNING: Only {len(unique_classes_in_data)} unique class(es) found ({unique_classes_in_data[0]}).")
    print("This pipeline is designed for binary classification. Proceeding with a single-class detection interpretation.")
    print("The existing class will be treated as the 'positive' class (label 1).")

    class_A = unique_classes_in_data[0] # The only class
    class_B = "NO_OTHER_CLASS" # Placeholder for reporting, won't be used in data processing
    label_map = {class_A: 1} # Map the single class to 1
    df2 = df[df["label"] == class_A].copy() # Ensure only this class is in df2
    df2["y"] = df2["label"].map(label_map).astype(int)

    stratify_split_param = None # Cannot stratify with only one class value
    class_weights_to_use = None # Not meaningful for single class

    print("\nSelected class for single-class detection:")
    print(f"Class (mapped to Y=1): {class_A}")

else:
    # Original binary classification logic
    top2 = unique_classes_in_data[:2]
    class_A, class_B = top2[0], top2[1]
    label_map = {class_A: 0, class_B: 1} # Map top 1 to 0, top 2 to 1 for binary
    df2 = df[df["label"].isin(top2)].copy()
    df2["y"] = df2["label"].map(label_map).astype(int)

    stratify_split_param = df2["y"]
    class_weights_to_use = True # Indicate that weights should be computed later


    print("\nSelected top-2 classes for binary:")
    print(f"class_A (mapped to Y=0): {class_A}")
    print(f"class_B (mapped to Y=1): {class_B}")


print("\nCounts in selected subset:\n", df2["label"].value_counts())
print("Total selected images:", len(df2))
print("Class ratios (y=1):", df2["y"].mean())

# -------------------------
# 2) STRATIFIED SPLIT (70/15/15) - Adapted
# -------------------------
if is_single_class: # Single class scenario
    train_df, temp_df = train_test_split(
        df2, test_size=0.30, random_state=SEED # No stratify
    )
    val_df, test_df = train_test_split(
        temp_df, test_size=0.50, random_state=SEED # No stratify
    )
else: # Binary class scenario
    train_df, temp_df = train_test_split(
        df2, test_size=0.30, stratify=stratify_split_param, random_state=SEED
    )
    val_df, test_df = train_test_split(
        temp_df, test_size=0.50, stratify=temp_df["y"], random_state=SEED
    )

def split_report(name, d):
    report_dict = {"n": len(d)}
    if is_single_class:
        report_dict[f"{class_A} (1)"] = int((d["y"]==1).sum())
        report_dict["pos_ratio(y=1)"] = float((d["y"]==1).mean())
    else:
        report_dict[f"{class_A} (0)"] = int((d["y"]==0).sum())
        report_dict[f"{class_B} (1)"] = int((d["y"]==1).sum())
        report_dict["pos_ratio(y=1)"] = float((d["y"]==1).mean())
    return pd.Series(report_dict, name=name)

report_table = pd.concat([
    split_report("train", train_df),
    split_report("val", val_df),
    split_report("test", test_df),
], axis=1).T

print("\nSplit distribution table:\n")
display(report_table)

# -------------------------
# Optional: class_weight (recommended if imbalance) - Adapted
# -------------------------
if is_single_class:
    print("\nClass weights not computed for single-class scenario (all samples are positive).")
    class_weight = None
else:
    y_train = train_df["y"].values
    classes = np.array([0, 1])
    cw = compute_class_weight(class_weight="balanced", classes=classes, y=y_train)
    class_weight = {0: float(cw[0]), 1: float(cw[1])}
    print("\nClass weights: ", class_weight)

# -------------------------
# 3) tf.data PIPELINE + TRAIN-ONLY AUGMENTATION
# -------------------------
def load_image(path, label):
    img = tf.io.read_file(path)
    # decode_image handles jpg/png but shape can be dynamic; we set shape later by resize
    img = tf.image.decode_image(img, channels=3, expand_animations=False)
    img = tf.image.resize(img, IMG_SIZE)
    img = tf.cast(img, tf.float32) / 255.0
    return img, label

data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal"),
    tf.keras.layers.RandomFlip("vertical"),
    tf.keras.layers.RandomRotation(0.1),   # ~±18°
    tf.keras.layers.RandomZoom(0.1),
    tf.keras.layers.RandomContrast(0.1),
], name="aug")

def make_dataset(df_in, training=False):
    ds = tf.data.Dataset.from_tensor_slices((df_in["path"].values, df_in["y"].values))
    if training:
        ds = ds.shuffle(2000, seed=SEED, reshuffle_each_iteration=True)
    ds = ds.map(load_image, num_parallel_calls=AUTOTUNE)
    if training:
        ds = ds.map(lambda x,y: (data_augmentation(x, training=True), y),
                    num_parallel_calls=AUTOTUNE)
    ds = ds.batch(BATCH_SIZE).cache().prefetch(AUTOTUNE)
    return ds

train_ds = make_dataset(train_df, training=True)
val_ds   = make_dataset(val_df, training=False)
test_ds  = make_dataset(test_df, training=False)

def show_aug_examples(ds, n=6):
    imgs, labels = next(iter(ds.unbatch().batch(n)))
    plt.figure(figsize=(12,6))
    for i in range(n):
        plt.subplot(2, n//2, i+1)
        plt.imshow(imgs[i])
        plt.title(f"y={int(labels[i])}")
        plt.axis("off")
    plt.suptitle("Augmentation Examples (Train Only)")
    plt.show()

show_aug_examples(train_ds, n=6)

# -------------------------
# 4) MODEL-1: SCRATCH CNN
# -------------------------
from tensorflow.keras import layers as L

def build_scratch_cnn():
    inp = L.Input(shape=(*IMG_SIZE, 3))
    x = inp

    # Block-1
    x = L.Conv2D(32, 3, padding="same")(x); x = L.BatchNormalization()(x); x = L.ReLU()(x)
    x = L.Conv2D(32, 3, padding="same")(x); x = L.BatchNormalization()(x); x = L.ReLU()(x)
    x = L.MaxPool2D()(x); x = L.Dropout(0.25)(x)

    # Block-2
    x = L.Conv2D(64, 3, padding="same")(x); x = L.BatchNormalization()(x); x = L.ReLU()(x)
    x = L.Conv2D(64, 3, padding="same")(x); x = L.BatchNormalization()(x); x = L.ReLU()(x)
    x = L.MaxPool2D()(x); x = L.Dropout(0.25)(x)

    # Block-3
    x = L.Conv2D(128, 3, padding="same")(x); x = L.BatchNormalization()(x); x = L.ReLU()(x)
    x = L.Conv2D(128, 3, padding="same")(x); x = L.BatchNormalization()(x); x = L.ReLU()(x)
    x = L.MaxPool2D()(x); x = L.Dropout(0.30)(x)

    # Block-4 (optional, included)
    x = L.Conv2D(256, 3, padding="same")(x); x = L.BatchNormalization()(x); x = L.ReLU()(x)
    x = L.MaxPool2D()(x); x = L.Dropout(0.35)(x)

    # Head
    x = L.GlobalAveragePooling2D()(x)
    x = L.Dense(128, activation="relu")(x)
    x = L.Dropout(0.5)(x)
    out = L.Dense(1, activation="sigmoid")(x)

    return tf.keras.Model(inp, out, name="ScratchCNN")

def compile_bin(model, lr):
    model.compile(
        optimizer=tf.keras.optimizers.Adam(lr),
        loss=tf.keras.losses.BinaryCrossentropy(),
        metrics=[tf.keras.metrics.BinaryAccuracy(name="acc"),
                 tf.keras.metrics.AUC(name="auc")]
    )

cb = [
    tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=10, restore_best_weights=True),
    tf.keras.callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.2, patience=5, min_lr=1e-6),
]

def plot_history(hist, title=""):
    h = hist.history
    plt.figure()
    plt.plot(h["loss"], label="train_loss")
    plt.plot(h["val_loss"], label="val_loss")
    plt.title(title + " Loss"); plt.legend(); plt.show()

    plt.figure()
    plt.plot(h["acc"], label="train_acc")
    plt.plot(h["val_acc"], label="val_acc")
    plt.title(title + " Accuracy"); plt.legend(); plt.show()

scratch = build_scratch_cnn()
compile_bin(scratch, 1e-3)
print("\nScratch CNN Summary:\n")
scratch.summary()

hist_scratch = scratch.fit(
    train_ds, validation_data=val_ds, epochs=100,
    callbacks=cb, class_weight=class_weight
)
plot_history(hist_scratch, "ScratchCNN")

# -------------------------
# 5) TRANSFER LEARNING (MobileNetV2, EfficientNetB0)
# -------------------------
def build_tl_model(base, name):
    base.trainable = False
    inp = L.Input(shape=(*IMG_SIZE, 3))
    x = inp
    x = base(x, training=False)
    x = L.GlobalAveragePooling2D()(x)
    x = L.Dropout(0.4)(x)
    out = L.Dense(1, activation="sigmoid")(x)
    return tf.keras.Model(inp, out, name=name)

def fine_tune_last_quarter(base_model, keep_bn_frozen=True):
    N = len(base_model.layers)
    fine_tune_from = int(0.75 * N)
    for i, layer in enumerate(base_model.layers):
        if i < fine_tune_from:
            layer.trainable = False
        else:
            if keep_bn_frozen and isinstance(layer, tf.keras.layers.BatchNormalization):
                layer.trainable = False
            else:
                layer.trainable = True
    return fine_tune_from, N

# --- MobileNetV2
mobilenet_base = tf.keras.applications.MobileNetV2(
    include_top=False, weights="imagenet", input_shape=(*IMG_SIZE, 3)
)
mob = build_tl_model(mobilenet_base, "MobileNetV2_FE")
compile_bin(mob, 1e-3)

print("\nTraining MobileNetV2 - Feature Extraction (freeze)...")
hist_mob_fe = mob.fit(train_ds, validation_data=val_ds, epochs=100, callbacks=cb,
                      class_weight=class_weight)
plot_history(hist_mob_fe, "MobileNetV2 FE")

ft_from, N = fine_tune_last_quarter(mobilenet_base, keep_bn_frozen=True)
print(f"MobileNetV2 fine-tune from layer {ft_from}/{N} (last 25% opened, BN frozen)")

compile_bin(mob, 1e-5)
print("\nTraining MobileNetV2 - Fine Tuning...")
hist_mob_ft = mob.fit(train_ds, validation_data=val_ds, epochs=100, callbacks=cb,
                      class_weight=class_weight)
plot_history(hist_mob_ft, "MobileNetV2 FT")

# --- EfficientNetB0
eff_base = tf.keras.applications.EfficientNetB0(
    include_top=False, weights="imagenet", input_shape=(*IMG_SIZE, 3)
)
eff = build_tl_model(eff_base, "EffNetB0_FE")
compile_bin(eff, 1e-3)

print("\nTraining EfficientNetB0 - Feature Extraction (freeze)...")
hist_eff_fe = eff.fit(train_ds, validation_data=val_ds, epochs=100, callbacks=cb,
                      class_weight=class_weight)
plot_history(hist_eff_fe, "EffNetB0 FE")

ft_from, N = fine_tune_last_quarter(eff_base, keep_bn_frozen=True)
print(f"EffNetB0 fine-tune from layer {ft_from}/{N} (last 25% opened, BN frozen)")

compile_bin(eff, 1e-5)
print("\nTraining EfficientNetB0 - Fine Tuning...")
hist_eff_ft = eff.fit(train_ds, validation_data=val_ds, epochs=100, callbacks=cb,
                      class_weight=class_weight)
plot_history(hist_eff_ft, "EffNetB0 FT")

# -------------------------
# 6) FINAL TEST EVALUATION (Acc/Prec/Rec/F1/ROC-AUC) + CM + ROC CURVES
# -------------------------
def predict_probs(model, ds):
    return model.predict(ds).ravel()

def eval_binary(y_true, probs, thr=0.5):
    y_pred = (probs >= thr).astype(int)
    results = {}
    results["Accuracy"] = accuracy_score(y_true, y_pred)

    # Check if there's enough variability in y_true or y_pred for binary metrics
    if is_single_class or len(np.unique(y_true)) < 2 or len(np.unique(y_pred)) < 2:
        results["Precision"] = np.nan
        results["Recall"] = np.nan
        results["F1"] = np.nan
        results["ROC_AUC"] = np.nan
        results["CM"] = "Not applicable for single-class or constant prediction."
        results["Report"] = "Binary classification metrics not applicable for single-class or constant prediction."
    else:
        results["Precision"] = precision_score(y_true, y_pred, zero_division=0)
        results["Recall"] = recall_score(y_true, y_pred, zero_division=0)
        results["F1"] = f1_score(y_true, y_pred, zero_division=0)
        results["ROC_AUC"] = roc_auc_score(y_true, probs)
        results["CM"] = confusion_matrix(y_true, y_pred)
        try:
            results["Report"] = classification_report(y_true, y_pred, target_names=[class_A, class_B], zero_division=0)
        except ValueError:
            results["Report"] = "Classification report not applicable for this prediction distribution."

    return results

def plot_cm(cm, title="Confusion Matrix"):
    if isinstance(cm, str): # if it's a message like "Not applicable..."
        print(cm)
        return
    plt.figure()
    plt.imshow(cm)
    plt.title(title)
    # Only set ticks if it's a valid 2x2 matrix and not a single-class scenario
    if not is_single_class and cm.shape == (2,2):
        plt.xticks([0,1],[class_A, class_B], rotation=45)
        plt.yticks([0,1],[class_A, class_B])
    else:
        # For single class or non-standard CM, just show default ticks or print a message
        pass
    for i,j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j,i,cm[i,j],ha="center",va="center")
    plt.xlabel("Pred"); plt.ylabel("True")
    plt.show()

y_test = test_df["y"].values

models = {
    "ScratchCNN": scratch,
    "MobileNetV2": mob,
    "EffNetB0": eff
}

results = {}
for name, m in models.items():
    probs = predict_probs(m, test_ds)
    results[name] = eval_binary(y_test, probs, thr=0.5)
    print("\n======================")
    print("MODEL:", name)
    print(results[name]["Report"])
    print("Confusion Matrix:\n", results[name]["CM"])
    plot_cm(results[name]["CM"], title=f"{name} - Confusion Matrix")

# Comparison table
comp = pd.DataFrame([
    {**{"Model": k},
     **{m: results[k][m] for m in ["Accuracy","Precision","Recall","F1","ROC_AUC"]}}
    for k in results.keys()
]).set_index("Model")
print("\n=== Test Metrics Comparison ===\n")
display(comp)

# ROC curves - Adapted for single class
plt.figure()
can_plot_roc = True
if is_single_class or len(np.unique(y_test)) < 2:
    print("\nSkipping ROC curve plotting: Not applicable for single class in test set.")
    can_plot_roc = False

if can_plot_roc:
    for name, m in models.items():
        probs = predict_probs(m, test_ds)
        # Check if probs also has variability or if model always predicts same thing
        # roc_auc_score and roc_curve require at least two classes (or unique prediction values)
        if len(np.unique(probs >= 0.5)) < 2 or len(np.unique(y_test)) < 2:
            print(f"Skipping ROC for {name}: Model output or true labels are constant on test set.")
            continue
        try:
            fpr, tpr, _ = roc_curve(y_test, probs)
            auc = roc_auc_score(y_test, probs)
            plt.plot(fpr, tpr, label=f"{name} (AUC={auc:.3f})")
        except ValueError as e:
            print(f"Skipping ROC for {name} due to error: {e}")
            continue
    plt.plot([0,1],[0,1],"--")
    plt.xlabel("FPR"); plt.ylabel("TPR"); plt.title("ROC Curves (Test)")
    plt.legend(); plt.show()

# -------------------------
# 7) GRAD-CAM (3 correct + 3 wrong if available)
# -------------------------
def get_last_conv_layer(model):
    # Finds last Conv2D in top-level; if none, user must set manually.
    for layer in reversed(model.layers):
        if isinstance(layer, tf.keras.layers.Conv2D):
            return layer.name
    # Try to search inside nested models (common in TL)
    for layer in reversed(model.layers):
        if isinstance(layer, tf.keras.Model):
            for sub in reversed(layer.layers):
                if isinstance(sub, tf.keras.layers.Conv2D):
                    return sub.name
    raise ValueError("Conv2D layer not found automatically. Pick a conv layer name from model.summary().")

def make_gradcam_heatmap(img_tensor, model, last_conv_name):
    # Build grad model
    try:
        conv_layer = model.get_layer(last_conv_name)
    except:
        # If the conv layer is inside a nested base model, locate it
        conv_layer = None
        for layer in model.layers:
            if isinstance(layer, tf.keras.Model):
                try:
                    conv_layer = layer.get_layer(last_conv_name)
                    break
                except:
                    pass
        if conv_layer is None:
            raise

    grad_model = tf.keras.Model([model.inputs], [conv_layer.output, model.output])

    with tf.GradientTape() as tape:
        conv_out, preds = grad_model(img_tensor)
        loss = preds[:, 0]

    grads = tape.gradient(loss, conv_out)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    conv_out = conv_out[0]
    heatmap = tf.reduce_sum(conv_out * pooled_grads, axis=-1) # <-- Added this line, as it was missing
    heatmap = tf.maximum(heatmap, 0) / (tf.reduce_max(heatmap) + 1e-8)
    return heatmap.numpy()

def overlay_heatmap(img, heatmap, alpha=0.35):
    import cv2
    h = cv2.resize(heatmap, (img.shape[1], img.shape[0]))
    h = np.uint8(255*h)
    h = cv2.applyColorMap(h, cv2.COLORMAP_JET)
    img_bgr = (img*255).astype(np.uint8)[:, :, ::-1]
    overlay = cv2.addWeighted(img_bgr, 1-alpha, h, alpha, 0)
    return overlay[:, :, ::-1]

def get_examples_for_gradcam(model, df_source, n_correct=3, n_wrong=3):
    correct, wrong = [], []
    # shuffle
    df_shuf = df_source.sample(frac=1, random_state=SEED).reset_index(drop=True)
    for _, row in df_shuf.iterrows():
        img_path, y = row["path"], int(row["y"])
        img, _ = load_image(img_path, y)
        img_tensor = tf.expand_dims(img, 0)
        p = float(model.predict(img_tensor, verbose=0)[0,0])
        yhat = int(p >= 0.5)
        item = (img.numpy(), y, p, yhat, img_tensor)
        if yhat == y and len(correct) < n_correct:
            correct.append(item)
        elif yhat != y and len(wrong) < n_wrong:
            wrong.append(item)
        if len(correct) == n_correct and len(wrong) == n_wrong:
            break
    return correct, wrong

def show_gradcam_set(model, model_name):
    print("\n======================")
    print("Grad-CAM for:", model_name)
    try:
        last_conv = get_last_conv_layer(model)
    except Exception as e:
        print("Could not auto-find last conv layer:", e)
        print("Tip: run model.summary() and set last_conv manually.")
        return

    print("Using last conv layer:", last_conv)
    corr, wr = get_examples_for_gradcam(model, test_df, 3, 3)

    def plot_items(items, title_prefix):
        for i,(img, y, p, yhat, img_tensor) in enumerate(items):
            heat = make_gradcam_heatmap(img_tensor, model, last_conv)
            over = overlay_heatmap(img, heat)

            plt.figure(figsize=(10,3))
            plt.subplot(1,3,1); plt.imshow(img);  plt.title(f"Orig y={y}"); plt.axis("off")
            plt.subplot(1,3,2); plt.imshow(heat); plt.title("Heatmap"); plt.axis("off")
            plt.subplot(1,3,3); plt.imshow(over); plt.title(f"Overlay p={p:.3f} yhat={yhat}"); plt.axis("off")
            plt.suptitle(f"{model_name} - {title_prefix} #{i+1}")
            plt.show()

    if len(corr) > 0:
        plot_items(corr, "Correct")
    else:
        print("No correct examples found (unexpected).")

    if len(wr) > 0:
        plot_items(wr, "Wrong")
    else:
        print("No wrong examples found (model may be perfect on sampled subset).")

# Need cv2 for overlay
try:
    import cv2
except:
    !pip -q install opencv-python
    import cv2

show_gradcam_set(scratch, "ScratchCNN")
show_gradcam_set(mob, "MobileNetV2")
show_gradcam_set(eff, "EffNetB0")

print("\nDONE ✅")
```
