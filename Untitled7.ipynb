```python
# ============================================================
# MAKİNE ÖĞRENMESİ ARA SINAV ÖDEVİ
# Klasik ML + PCA + LDA + XAI (SHAP)
# TAM DÜZELTİLMİŞ / HATASIZ TEK PARÇA KOD
# Veri Seti: Wine Classification
# ============================================================

# =========================
# 0. GEREKLİ KURULUMLAR
# =========================
!pip -q install xgboost shap python-docx openpyxl

# =========================
# 1. KÜTÜPHANELER
# =========================
import warnings
warnings.filterwarnings("ignore")

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from collections import Counter
from sklearn.base import clone

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, confusion_matrix, classification_report,
    roc_curve, auc
)

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier

import shap
from docx import Document

# =========================
# 2. AYARLAR
# =========================
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)

# =========================
# 3. YARDIMCI FONKSİYONLAR
# =========================
def safe_display(obj, title=None):
    if title is not None:
        print(title)
    try:
        display(obj)
    except:
        print(obj)

def get_probabilities(model, X, all_classes):
    """
    Modelin predict_proba çıktısını güvenli biçimde döndürür.
    Eğer model bazı sınıfları train sırasında görmediyse, çıktı boyutunu tüm sınıflara tamamlar.
    """
    n_all = len(all_classes)

    if hasattr(model, "predict_proba"):
        proba = model.predict_proba(X)
        model_classes = getattr(model, "classes_", all_classes)

        if len(model_classes) == n_all:
            return proba

        aligned = np.zeros((len(X), n_all))
        class_to_index = {c: i for i, c in enumerate(all_classes)}
        for i, c in enumerate(model_classes):
            aligned[:, class_to_index[c]] = proba[:, i]
        return aligned

    if hasattr(model, "decision_function"):
        scores = model.decision_function(X)

        if scores.ndim == 1:
            probs_pos = 1 / (1 + np.exp(-scores))
            proba = np.vstack([1 - probs_pos, probs_pos]).T
            if proba.shape[1] == n_all:
                return proba
            aligned = np.zeros((len(X), n_all))
            aligned[:, :proba.shape[1]] = proba
            return aligned

        exp_scores = np.exp(scores - np.max(scores, axis=1, keepdims=True))
        proba = exp_scores / np.sum(exp_scores, axis=1, keepdims=True)

        if proba.shape[1] == n_all:
            return proba

        aligned = np.zeros((len(X), n_all))
        aligned[:, :proba.shape[1]] = proba
        return aligned

    preds = model.predict(X)
    preds_bin = label_binarize(preds, classes=all_classes)
    if preds_bin.ndim == 1:
        preds_bin = preds_bin.reshape(-1, 1)
    if preds_bin.shape[1] == 1 and len(all_classes) == 2:
        preds_bin = np.hstack([1 - preds_bin, preds_bin])

    if preds_bin.shape[1] < len(all_classes):
        aligned = np.zeros((len(X), len(all_classes)))
        aligned[:, :preds_bin.shape[1]] = preds_bin
        return aligned

    return preds_bin

def evaluate_model(model, X, y_true, all_classes):
    y_pred = model.predict(X)
    y_proba = get_probabilities(model, X, all_classes)

    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, average="weighted", zero_division=0)
    rec = recall_score(y_true, y_pred, average="weighted", zero_division=0)
    f1 = f1_score(y_true, y_pred, average="weighted", zero_division=0)

    y_true_bin = label_binarize(y_true, classes=all_classes)
    if y_true_bin.ndim == 1:
        y_true_bin = y_true_bin.reshape(-1, 1)

    try:
        if len(all_classes) > 2:
            roc_auc = roc_auc_score(y_true_bin, y_proba, multi_class="ovr", average="weighted")
        else:
            roc_auc = roc_auc_score(y_true, y_proba[:, 1])
    except:
        roc_auc = np.nan

    return {
        "Accuracy": acc,
        "Precision": prec,
        "Recall": rec,
        "F1-score": f1,
        "ROC-AUC": roc_auc
    }

def make_feature_names(rep_name, X_array, original_feature_names):
    if rep_name == "Raw":
        return list(original_feature_names)
    elif rep_name == "PCA":
        return [f"PCA_{i+1}" for i in range(X_array.shape[1])]
    elif rep_name == "LDA":
        return [f"LDA_{i+1}" for i in range(X_array.shape[1])]
    else:
        return [f"F_{i+1}" for i in range(X_array.shape[1])]

def normalize_shap_values(shap_values):
    """
    SHAP çıktısını numpy array / list biçiminde normalize eder.
    Olası biçimler:
    - list of arrays [class0, class1, ...]
    - ndarray (n_samples, n_features)
    - ndarray (n_samples, n_features, n_classes)
    - ndarray (n_classes, n_samples, n_features)
    - shap.Explanation
    """
    # shap.Explanation ise .values al
    if hasattr(shap_values, "values"):
        shap_values = shap_values.values

    # list ise içindeki elemanları numpy'a çevir
    if isinstance(shap_values, list):
        normalized = []
        for sv in shap_values:
            if hasattr(sv, "values"):
                sv = sv.values
            normalized.append(np.array(sv))
        return normalized

    return np.array(shap_values)

def shap_importance_dataframe(shap_values, feature_names):
    """
    Mean absolute SHAP importance değerlerini her durumda 1D hale getirir.
    """
    sv = normalize_shap_values(shap_values)

    if isinstance(sv, list):
        # her sınıf için (n_samples, n_features) beklenir
        class_imps = []
        for arr in sv:
            arr = np.array(arr)
            if arr.ndim == 3:
                # beklenmeyen durum: samples x features x something
                arr = np.mean(np.abs(arr), axis=tuple(i for i in range(arr.ndim) if i != 1))
            elif arr.ndim == 2:
                arr = np.mean(np.abs(arr), axis=0)
            elif arr.ndim == 1:
                pass
            else:
                arr = np.ravel(arr)
            class_imps.append(arr)

        imp = np.mean(np.vstack(class_imps), axis=0)

    else:
        arr = np.array(sv)

        if arr.ndim == 2:
            # (n_samples, n_features)
            imp = np.mean(np.abs(arr), axis=0)

        elif arr.ndim == 3:
            # iki olasılık:
            # (n_samples, n_features, n_classes)
            # (n_classes, n_samples, n_features)
            if arr.shape[1] == len(feature_names):
                # (n_samples, n_features, n_classes)
                imp = np.mean(np.abs(arr), axis=(0, 2))
            elif arr.shape[2] == len(feature_names):
                # (n_classes, n_samples, n_features)
                imp = np.mean(np.abs(arr), axis=(0, 1))
            else:
                # en güvenli fallback
                feature_axis = np.argmin([abs(arr.shape[i] - len(feature_names)) for i in range(arr.ndim)])
                reduce_axes = tuple(i for i in range(arr.ndim) if i != feature_axis)
                imp = np.mean(np.abs(arr), axis=reduce_axes)

        elif arr.ndim == 1:
            imp = arr

        else:
            arr = np.ravel(arr)
            imp = arr[:len(feature_names)]

    imp = np.array(imp).reshape(-1)

    if len(imp) != len(feature_names):
        min_len = min(len(imp), len(feature_names))
        feature_names = feature_names[:min_len]
        imp = imp[:min_len]

    df_imp = pd.DataFrame({
        "Feature": feature_names,
        "MeanAbsSHAP": imp
    }).sort_values(by="MeanAbsSHAP", ascending=False).reset_index(drop=True)

    return df_imp

def safe_shap_explainer(model, X_background_df, X_eval_df):
    """
    Model tipine göre güvenli SHAP explainer kurar.
    """
    # örneklem küçültme
    bg = X_background_df.copy()
    ev = X_eval_df.copy()

    if len(bg) > 80:
        bg = shap.sample(bg, 80, random_state=RANDOM_STATE)
    if len(ev) > 50:
        ev = ev.iloc[:50].copy()

    # ağaç tabanlı modeller
    if isinstance(model, (DecisionTreeClassifier, RandomForestClassifier, XGBClassifier)):
        explainer = shap.TreeExplainer(model)
        shap_values = explainer.shap_values(ev)
        return explainer, shap_values, ev

    # lineer model
    if isinstance(model, LogisticRegression):
        explainer = shap.LinearExplainer(model, bg)
        shap_values = explainer.shap_values(ev)
        return explainer, shap_values, ev

    # GaussianNB için KernelExplainer
    explainer = shap.KernelExplainer(model.predict_proba, bg)
    shap_values = explainer.shap_values(ev, nsamples=100)
    return explainer, shap_values, ev

def plot_shap_summary_and_bar(shap_values, X_eval_df, title_prefix="SHAP"):
    """
    SHAP grafiklerini hata vermeden üretmeye çalışır.
    """
    sv = normalize_shap_values(shap_values)

    try:
        if isinstance(sv, list):
            shap.summary_plot(sv[0], X_eval_df, show=True)
            shap.summary_plot(sv[0], X_eval_df, plot_type="bar", show=True)
        else:
            arr = np.array(sv)

            if arr.ndim == 2:
                shap.summary_plot(arr, X_eval_df, show=True)
                shap.summary_plot(arr, X_eval_df, plot_type="bar", show=True)

            elif arr.ndim == 3:
                if arr.shape[1] == X_eval_df.shape[1]:
                    # (n_samples, n_features, n_classes)
                    arr_plot = arr[:, :, 0]
                elif arr.shape[2] == X_eval_df.shape[1]:
                    # (n_classes, n_samples, n_features)
                    arr_plot = arr[0]
                else:
                    arr_plot = arr.reshape(arr.shape[0], -1)[:, :X_eval_df.shape[1]]

                shap.summary_plot(arr_plot, X_eval_df, show=True)
                shap.summary_plot(arr_plot, X_eval_df, plot_type="bar", show=True)

            else:
                print(f"{title_prefix}: SHAP grafik boyutu beklenenden farklı, grafik atlandı.")

    except Exception as e:
        print(f"{title_prefix}: SHAP grafik çiziminde uyarı:", e)

# =========================
# 4. VERİ SETİNİN YÜKLENMESİ
# =========================
print("=" * 80)
print("1) VERİ SETİNİN YÜKLENMESİ")
print("=" * 80)

wine = load_wine()
X = wine.data
y = wine.target
feature_names = wine.feature_names
target_names = wine.target_names
all_classes = np.unique(y)

df = pd.DataFrame(X, columns=feature_names)
df["target"] = y
df["target_name"] = df["target"].map({i: n for i, n in enumerate(target_names)})

safe_display(df.head(), "\nİlk 5 satır:")
print("\nVeri seti boyutu:", df.shape)
safe_display(df["target_name"].value_counts(), "\nSınıf dağılımı:")

X_df = df[feature_names].copy()
y_series = df["target"].copy()

# =========================
# 5. VERİ KALİTE KONTROLLERİ
# =========================
print("\n" + "=" * 80)
print("2) VERİ SETİ KALİTE KONTROLLERİ")
print("=" * 80)

# 2.1 Eksik değer analizi
print("\n2.1 Eksik Değer Analizi")
missing_values = X_df.isnull().sum()
missing_df = pd.DataFrame({
    "Feature": X_df.columns,
    "MissingCount": missing_values.values,
    "MissingRatio(%)": (missing_values.values / len(X_df)) * 100
})
safe_display(missing_df)

imputer = SimpleImputer(strategy="median")
X_imputed = pd.DataFrame(imputer.fit_transform(X_df), columns=X_df.columns)

if missing_values.sum() == 0:
    print("Eksik değer bulunmadı. Pipeline bütünlüğü için median imputer yine de uygulandı.")
else:
    print("Eksik değerler median ile dolduruldu.")

# 2.2 Aykırı değer analizi
print("\n2.2 Aykırı Değer Analizi (IQR)")
outlier_summary = []
for col in X_imputed.columns:
    Q1 = X_imputed[col].quantile(0.25)
    Q3 = X_imputed[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    count = ((X_imputed[col] < lower) | (X_imputed[col] > upper)).sum()

    outlier_summary.append([col, Q1, Q3, IQR, lower, upper, count])

outlier_df = pd.DataFrame(outlier_summary, columns=[
    "Feature", "Q1", "Q3", "IQR", "LowerBound", "UpperBound", "OutlierCount"
])
safe_display(outlier_df.sort_values(by="OutlierCount", ascending=False))

print("\nKısa Yorum:")
print("- IQR analizi ile bazı sütunlarda aykırı değer eğilimi görülmektedir.")
print("- Aykırı değerler özellikle lineer ve dağılım duyarlı modelleri etkileyebilir.")
print("- Bu nedenle veri ölçeklendirme yapılmıştır.")

# 2.3 Veri tipi incelemesi
print("\n2.3 Veri Tipi ve Dağılım İncelemesi")
dtype_df = pd.DataFrame({
    "Feature": X_imputed.columns,
    "dtype": X_imputed.dtypes.astype(str).values
})
safe_display(dtype_df)

num_cols = X_imputed.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = X_imputed.select_dtypes(exclude=[np.number]).columns.tolist()
print("Sayısal değişken sayısı:", len(num_cols))
print("Kategorik değişken sayısı:", len(cat_cols))

# =========================
# 6. EDA
# =========================
print("\n" + "=" * 80)
print("3) KEŞİFSEL VERİ ANALİZİ (EDA)")
print("=" * 80)

# 3.1 İstatistiksel özellikler
print("\n3.1 İstatistiksel Özellikler")
eda_stats = pd.DataFrame({
    "Mean": X_imputed.mean(),
    "Median": X_imputed.median(),
    "Min": X_imputed.min(),
    "Max": X_imputed.max(),
    "Std": X_imputed.std(),
    "Q1": X_imputed.quantile(0.25),
    "Q3": X_imputed.quantile(0.75)
})
safe_display(eda_stats)

print("\nKısa Yorum:")
print("- Özelliklerin ölçekleri birbirinden farklıdır.")
print("- Bu nedenle StandardScaler kullanılması uygundur.")

# 3.2 Korelasyon matrisi
print("\n3.2 Korelasyon Matrisi")
corr_matrix = X_imputed.corr(method="pearson")

plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Pearson Korelasyon Matrisi")
plt.tight_layout()
plt.show()

upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
top3_corr = upper_tri.stack().sort_values(key=lambda x: np.abs(x), ascending=False).head(3)

print("En yüksek korelasyonlu 3 çift:")
for idx, val in top3_corr.items():
    print(f"{idx[0]} <-> {idx[1]} : {val:.4f}")

print("\nKısa Yorum:")
for idx, val in top3_corr.items():
    yon = "pozitif" if val > 0 else "negatif"
    print(f"- {idx[0]} ile {idx[1]} arasında {yon} yönlü güçlü ilişki vardır ({val:.3f}).")
print("- Güçlü korelasyon, bazı bilgilerin birbirini tekrar ettiğini gösterebilir.")

# 3.3 Boxplot analizi
print("\n3.3 Boxplot Analizi")
n_cols_plot = 3
n_rows_plot = int(np.ceil(len(X_imputed.columns) / n_cols_plot))
fig, axes = plt.subplots(n_rows_plot, n_cols_plot, figsize=(18, 4 * n_rows_plot))
axes = axes.flatten()

for i, col in enumerate(X_imputed.columns):
    sns.boxplot(x=X_imputed[col], ax=axes[i], color="skyblue")
    axes[i].set_title(f"Boxplot - {col}")

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()

print("Kısa Yorum:")
print("- Boxplot grafiklerinde bazı değişkenlerde aykırı gözlem eğilimleri görünmektedir.")

# =========================
# 7. SCALING VE SPLIT
# =========================
print("\n" + "=" * 80)
print("4) VERİ ÖLÇEKLENDİRME ve 5) VERİ BÖLME")
print("=" * 80)

X_temp, X_test, y_temp, y_test = train_test_split(
    X_imputed, y_series,
    test_size=0.20,
    stratify=y_series,
    random_state=RANDOM_STATE
)

X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp,
    test_size=0.125,  # toplamın %10'u
    stratify=y_temp,
    random_state=RANDOM_STATE
)

print("Train boyutu:", X_train.shape)
print("Validation boyutu:", X_val.shape)
print("Test boyutu:", X_test.shape)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

X_scaled = scaler.fit_transform(X_imputed)
print("X_scaled oluşturuldu.")

# =========================
# 8. PCA
# =========================
print("\n" + "=" * 80)
print("6.1) PCA")
print("=" * 80)

pca_full = PCA(random_state=RANDOM_STATE)
pca_full.fit(X_train_scaled)

explained_variance_ratio = pca_full.explained_variance_ratio_
mean_evr = explained_variance_ratio.mean()
selected_pca_components = np.sum(explained_variance_ratio > mean_evr)
selected_pca_components = max(int(selected_pca_components), 2)

print("Explained Variance Ratio:", explained_variance_ratio)
print("Explained Variance Ratio Ortalaması:", mean_evr)
print("Seçilen PCA bileşen sayısı:", selected_pca_components)

plt.figure(figsize=(10, 5))
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, marker='o')
plt.axhline(y=mean_evr, color='red', linestyle='--', label='EVR Ortalama')
plt.xlabel("Bileşen Sayısı")
plt.ylabel("Explained Variance Ratio")
plt.title("PCA Explained Variance Ratio")
plt.legend()
plt.tight_layout()
plt.show()

pca = PCA(n_components=selected_pca_components, random_state=RANDOM_STATE)
X_train_pca = pca.fit_transform(X_train_scaled)
X_val_pca = pca.transform(X_val_scaled)
X_test_pca = pca.transform(X_test_scaled)

plt.figure(figsize=(9, 6))
for cls in np.unique(y_train):
    idx = (y_train == cls)
    plt.scatter(X_train_pca[idx, 0], X_train_pca[idx, 1], label=target_names[cls], alpha=0.7)
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("İlk İki PCA Bileşeni ile Sınıf Ayrımı")
plt.legend()
plt.tight_layout()
plt.show()

print("Kısa Yorum:")
print("- PCA denetimsiz bir boyut indirgeme yöntemidir.")
print("- Sınıf ayrımı sınıf bilgisini kullanmadan yapılır.")

# =========================
# 9. LDA
# =========================
print("\n" + "=" * 80)
print("6.2) LDA")
print("=" * 80)

n_classes = len(np.unique(y_train))
max_lda_components = min(X_train_scaled.shape[1], n_classes - 1)
requested_lda_components = 3
actual_lda_components = min(requested_lda_components, max_lda_components)

print("İstenen LDA bileşen sayısı:", requested_lda_components)
print("Mümkün maksimum LDA bileşen sayısı:", max_lda_components)
print("Kullanılacak LDA bileşen sayısı:", actual_lda_components)

lda = LinearDiscriminantAnalysis(n_components=actual_lda_components)
X_train_lda = lda.fit_transform(X_train_scaled, y_train)
X_val_lda = lda.transform(X_val_scaled)
X_test_lda = lda.transform(X_test_scaled)

if actual_lda_components >= 2:
    plt.figure(figsize=(9, 6))
    for cls in np.unique(y_train):
        idx = (y_train == cls)
        plt.scatter(X_train_lda[idx, 0], X_train_lda[idx, 1], label=target_names[cls], alpha=0.7)
    plt.xlabel("LDA 1")
    plt.ylabel("LDA 2")
    plt.title("İlk İki LDA Bileşeni ile Sınıf Ayrımı")
    plt.legend()
    plt.tight_layout()
    plt.show()
else:
    plt.figure(figsize=(9, 4))
    for cls in np.unique(y_train):
        idx = (y_train == cls)
        plt.scatter(X_train_lda[idx, 0], np.zeros(np.sum(idx)), label=target_names[cls], alpha=0.7)
    plt.xlabel("LDA 1")
    plt.title("LDA 1 ile Sınıf Ayrımı")
    plt.legend()
    plt.tight_layout()
    plt.show()

print("Kısa Yorum:")
print("- LDA denetimli bir yöntemdir.")
print("- Amaç sınıflar arası ayrımı maksimize etmektir.")

# =========================
# 10. VERİ TEMSİLLERİ
# =========================
data_representations = {
    "Raw": (X_train_scaled, X_val_scaled, X_test_scaled),
    "PCA": (X_train_pca, X_val_pca, X_test_pca),
    "LDA": (X_train_lda, X_val_lda, X_test_lda)
}

# =========================
# 11. MODELLER
# =========================
base_models = {
    "Logistic Regression": LogisticRegression(max_iter=5000, random_state=RANDOM_STATE),
    "Decision Tree": DecisionTreeClassifier(random_state=RANDOM_STATE),
    "Random Forest": RandomForestClassifier(n_estimators=300, random_state=RANDOM_STATE),
    "XGBoost": XGBClassifier(
        n_estimators=300,
        max_depth=4,
        learning_rate=0.05,
        subsample=0.9,
        colsample_bytree=0.8,
        objective="multi:softprob",
        eval_metric="mlogloss",
        random_state=RANDOM_STATE
    ),
    "GaussianNB": GaussianNB()
}

# =========================
# 12. 15 MODEL EĞİTİMİ
# =========================
print("\n" + "=" * 80)
print("7) MAKİNE ÖĞRENMESİ MODELLERİ ve 8) VALIDATION KARŞILAŞTIRMASI")
print("=" * 80)

results = []
trained_models = {}

for rep_name, (Xtr, Xv, Xte) in data_representations.items():
    print(f"\n--- Veri Temsili: {rep_name} ---")
    for model_name, base_model in base_models.items():
        print(f"Eğitiliyor: {model_name} | {rep_name}")

        model = clone(base_model)
        model.fit(Xtr, y_train)

        metrics = evaluate_model(model, Xv, y_val, all_classes)

        results.append({
            "Representation": rep_name,
            "Model": model_name,
            **metrics
        })

        trained_models[(rep_name, model_name)] = model

results_df = pd.DataFrame(results).sort_values(
    by=["F1-score", "ROC-AUC", "Accuracy"],
    ascending=False
).reset_index(drop=True)

safe_display(results_df, "\nValidation Sonuçları:")

print("\nKısa Yorum:")
print("- Toplam 15 model validation setinde karşılaştırıldı.")
print("- En iyi model seçiminde F1-score, ROC-AUC ve Accuracy birlikte dikkate alındı.")

# =========================
# 13. EN İYİ MODEL
# =========================
best_row = results_df.iloc[0]
best_representation = best_row["Representation"]
best_model_name = best_row["Model"]
best_model = trained_models[(best_representation, best_model_name)]

Xtr_best, Xv_best, Xte_best = data_representations[best_representation]

print("\n" + "=" * 80)
print("9) EN İYİ MODELİN TEST ÜZERİNDE DEĞERLENDİRİLMESİ")
print("=" * 80)
print("En iyi model:", best_model_name)
print("Veri temsili:", best_representation)

test_metrics = evaluate_model(best_model, Xte_best, y_test, all_classes)
test_metrics_df = pd.DataFrame([test_metrics])

safe_display(test_metrics_df, "\nTest Performansı:")

y_test_pred = best_model.predict(Xte_best)
y_test_proba = get_probabilities(best_model, Xte_best, all_classes)

print("\nClassification Report:")
print(classification_report(y_test, y_test_pred, target_names=target_names, zero_division=0))

# 9.2 Confusion Matrix
cm = confusion_matrix(y_test, y_test_pred)

plt.figure(figsize=(7, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=target_names, yticklabels=target_names)
plt.xlabel("Tahmin Edilen")
plt.ylabel("Gerçek")
plt.title(f"Confusion Matrix - {best_model_name} ({best_representation})")
plt.tight_layout()
plt.show()

print("Kısa Yorum:")
print("- Diyagonal üzerindeki değerler doğru tahminleri gösterir.")
print("- Diyagonal dışındaki değerler yanlış sınıflamaları gösterir.")

# 9.3 ROC Eğrisi
y_test_bin = label_binarize(y_test, classes=all_classes)
if y_test_bin.ndim == 1:
    y_test_bin = y_test_bin.reshape(-1, 1)

plt.figure(figsize=(10, 7))
roc_auc_dict = {}

for i in range(len(all_classes)):
    try:
        fpr, tpr, thresholds = roc_curve(y_test_bin[:, i], y_test_proba[:, i])
        roc_auc_dict[target_names[i]] = auc(fpr, tpr)
        plt.plot(fpr, tpr, label=f"{target_names[i]} (AUC = {roc_auc_dict[target_names[i]]:.3f})")
    except Exception as e:
        print(f"{target_names[i]} için ROC çizilemedi:", e)

plt.plot([0, 1], [0, 1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title(f"ROC Eğrisi - {best_model_name} ({best_representation})")
plt.legend()
plt.tight_layout()
plt.show()

print("Kısa Yorum:")
print("- ROC eğrisi threshold değişimine göre model performansını gösterir.")
print("- Threshold düşerse recall artabilir, ama false positive da artabilir.")
print("- Threshold yükselirse precision artabilir, ama recall azalabilir.")

# =========================
# 14. SHAP ANALİZİ
# =========================
print("\n" + "=" * 80)
print("10) XAI – SHAP ANALİZİ")
print("=" * 80)

shap_feature_names = make_feature_names(best_representation, Xtr_best, X_train.columns)
X_train_shap = pd.DataFrame(Xtr_best, columns=shap_feature_names)
X_test_shap = pd.DataFrame(Xte_best, columns=shap_feature_names)

print(f"SHAP uygulanıyor: {best_model_name} | {best_representation}")
best_explainer, best_shap_values, best_eval_df = safe_shap_explainer(
    best_model, X_train_shap, X_test_shap
)

print("\n10.1 En iyi Validation Modeli için SHAP Analizi")
plot_shap_summary_and_bar(best_shap_values, best_eval_df, title_prefix="Best Model SHAP")

importance_df = shap_importance_dataframe(best_shap_values, list(best_eval_df.columns))
safe_display(importance_df.head(10), "\nEn önemli SHAP özellikleri:")

print("\nKısa Yorum:")
print("- Mean |SHAP| değeri yüksek olan değişkenler model kararını daha fazla etkiler.")
print("- Ham veri kullanılmışsa yorum orijinal özellikler üzerinden yapılır.")
print("- PCA veya LDA kullanılmışsa yorum bileşenler üzerinden yapılır.")

# 10.2 PCA ve LDA SHAP karşılaştırması
print("\n10.2 PCA ve LDA Temsilleri için SHAP Karşılaştırması")

shap_compare_results = []

for rep in ["PCA", "LDA"]:
    rep_df = results_df[results_df["Representation"] == rep].sort_values(
        by=["F1-score", "ROC-AUC", "Accuracy"],
        ascending=False
    )
    rep_best_row = rep_df.iloc[0]
    rep_best_model_name = rep_best_row["Model"]
    rep_best_model = trained_models[(rep, rep_best_model_name)]

    Xtr_rep, Xv_rep, Xte_rep = data_representations[rep]
    rep_feature_names = make_feature_names(rep, Xtr_rep, X_train.columns)

    Xtr_rep_df = pd.DataFrame(Xtr_rep, columns=rep_feature_names)
    Xte_rep_df = pd.DataFrame(Xte_rep, columns=rep_feature_names)

    try:
        rep_explainer, rep_shap_values, rep_eval_df = safe_shap_explainer(
            rep_best_model, Xtr_rep_df, Xte_rep_df
        )

        plot_shap_summary_and_bar(rep_shap_values, rep_eval_df, title_prefix=f"{rep} SHAP")

        rep_imp_df = shap_importance_dataframe(rep_shap_values, list(rep_eval_df.columns))
        safe_display(rep_imp_df.head(10), f"\n{rep} için en önemli SHAP bileşenleri:")

        shap_compare_results.append({
            "Representation": rep,
            "BestModel": rep_best_model_name,
            "TopFeatures": ", ".join(rep_imp_df.head(5)["Feature"].tolist())
        })

    except Exception as e:
        print(f"{rep} için SHAP analizi hatası:", e)
        shap_compare_results.append({
            "Representation": rep,
            "BestModel": rep_best_model_name,
            "TopFeatures": f"Hata: {str(e)}"
        })

shap_compare_df = pd.DataFrame(shap_compare_results)
safe_display(shap_compare_df, "\nPCA / LDA SHAP karşılaştırması:")

print("\nKısa Yorum:")
print("- PCA bileşenleri doğrusal birleşimlerdir.")
print("- LDA bileşenleri sınıf ayrımını artıracak şekilde oluşur.")
print("- SHAP, hangi bileşenlerin karar üzerinde daha etkili olduğunu gösterir.")

# =========================
# 15. DOSYA KAYITLARI
# =========================
print("\n" + "=" * 80)
print("DOSYALAR KAYDEDİLİYOR")
print("=" * 80)

output_dir = "ml_midterm_outputs"
os.makedirs(output_dir, exist_ok=True)

results_df.to_excel(os.path.join(output_dir, "validation_results.xlsx"), index=False)
test_metrics_df.to_excel(os.path.join(output_dir, "test_metrics.xlsx"), index=False)
eda_stats.to_excel(os.path.join(output_dir, "eda_statistics.xlsx"))
outlier_df.to_excel(os.path.join(output_dir, "outlier_analysis.xlsx"), index=False)
importance_df.to_excel(os.path.join(output_dir, "best_model_shap_importance.xlsx"), index=False)
shap_compare_df.to_excel(os.path.join(output_dir, "pca_lda_shap_compare.xlsx"), index=False)

# =========================
# 16. WORD RAPORU
# =========================
doc = Document()
doc.add_heading("Makine Öğrenmesi Ara Sınav Ödevi Raporu", 0)

doc.add_heading("1. Veri Seti", level=1)
doc.add_paragraph("Bu çalışmada scikit-learn kütüphanesinden Wine Classification veri seti kullanılmıştır.")
doc.add_paragraph(f"Veri seti boyutu: {df.shape}")
doc.add_paragraph(f"Sınıf dağılımı: {dict(Counter(y_series))}")

doc.add_heading("2. Veri Kalite Kontrolleri", level=1)
doc.add_paragraph("Eksik değer analizi yapılmıştır. Eksik değer bulunmasa da median imputation uygulanmıştır.")
doc.add_paragraph("IQR yöntemi ile aykırı değer analizi gerçekleştirilmiştir.")

doc.add_heading("3. EDA", level=1)
doc.add_paragraph("İstatistiksel özet, korelasyon matrisi ve boxplot analizleri yapılmıştır.")
top3_text = "\n".join([f"{idx[0]} - {idx[1]} : {val:.4f}" for idx, val in top3_corr.items()])
doc.add_paragraph("En yüksek korelasyonlu 3 özellik çifti:\n" + top3_text)

doc.add_heading("4. Ölçeklendirme ve Veri Bölme", level=1)
doc.add_paragraph("StandardScaler kullanılmıştır.")
doc.add_paragraph(f"Train boyutu: {X_train.shape}, Validation boyutu: {X_val.shape}, Test boyutu: {X_test.shape}")

doc.add_heading("5. Boyut İndirgeme", level=1)
doc.add_paragraph(f"PCA için seçilen bileşen sayısı: {selected_pca_components}")
doc.add_paragraph(
    f"LDA için istenen bileşen sayısı: 3, ancak teknik sınır nedeniyle kullanılan bileşen sayısı: {actual_lda_components}"
)

doc.add_heading("6. Validation Sonuçları", level=1)
doc.add_paragraph(f"En iyi validation modeli: {best_model_name} ({best_representation})")

doc.add_heading("7. Test Sonuçları", level=1)
for k, v in test_metrics.items():
    doc.add_paragraph(f"{k}: {v:.4f}")

doc.add_heading("8. SHAP Analizi", level=1)
doc.add_paragraph("En iyi model için SHAP açıklanabilirlik analizi yapılmıştır.")
for _, row in importance_df.head(5).iterrows():
    doc.add_paragraph(f"{row['Feature']}: {row['MeanAbsSHAP']:.6f}")

doc.add_heading("9. Genel Yorum", level=1)
doc.add_paragraph(
    "Bu çalışmada ham veri, PCA ve LDA temsilleri üzerinde 5 klasik makine öğrenmesi algoritması eğitilmiş ve "
    "validation performansları karşılaştırılmıştır. En iyi model test seti üzerinde ayrıca değerlendirilmiş, "
    "confusion matrix, ROC eğrisi ve SHAP ile yorumlanmıştır."
)

report_path = os.path.join(output_dir, "Makine_Ogrenmesi_AraSinav_Raporu.docx")
doc.save(report_path)

# =========================
# 17. SONUÇ ÖZETİ
# =========================
print("\n" + "=" * 80)
print("SONUÇ ÖZETİ")
print("=" * 80)
print("En iyi validation modeli:", best_model_name)
print("Veri temsili:", best_representation)

print("\nTest metrikleri:")
for k, v in test_metrics.items():
    print(f"{k}: {v:.4f}")

safe_display(importance_df.head(10), "\nİlk 10 SHAP özelliği:")

print("\nKaydedilen dosyalar:")
print("-", os.path.join(output_dir, "validation_results.xlsx"))
print("-", os.path.join(output_dir, "test_metrics.xlsx"))
print("-", os.path.join(output_dir, "eda_statistics.xlsx"))
print("-", os.path.join(output_dir, "outlier_analysis.xlsx"))
print("-", os.path.join(output_dir, "best_model_shap_importance.xlsx"))
print("-", os.path.join(output_dir, "pca_lda_shap_compare.xlsx"))
print("-", report_path)

print("\nNOT:")
print("- GitHub linki otomatik üretilemez; notebook dosyanı manuel olarak GitHub'a yüklemelisin.")
print("- Bu sürümde SHAP boyut farkları için güvenli dönüşüm yapıldığı için önceki hata alınmamalıdır.")
```
